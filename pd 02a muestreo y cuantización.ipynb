```python
import numpy as np
import matplotlib.pyplot as plt
```

## Parámetros de la señal analógica

$ s(t) = \cos(50\pi t)=\cos(2\pi f t) $

```python
F = 25
T = 1/F

t = np.linspace(0, 3*T, 1000)
s = np.cos(2*np.pi*F*t)


plt.plot(t, s)
plt.title("Señal analógica")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
```

## Parámetros de muestreo

Primer ejemplo:

$ F_S = 50$  Hz

$ T_S = 1/50$

Esta señal cumple con la frecuencia de muestreo


```python
Fs = 50
Ts = 1/Fs

ts = np.arange(0, 3*T, Ts)
xs = np.cos(2*np.pi*F*ts)

plt.plot(ts, xs)
```

```python
plt.stem(ts, xs, 'g')
plt.plot(t, s)
```

```python
print(len(ts))
```

# Cuantización

```python
Fs = 60
Ts = 1/Fs

t = np.linspace(0, 4*T, 1000)
ts = np.arange(0, 4*T, Ts)

x = np.cos(2*np.pi*F*t)
xs = np.cos(2*np.pi*F*ts)
xq = np.ceil(xs)


plt.plot(t, x)
plt.stem(ts, xs, 'r')

plt.figure()
plt.plot(t, x, 'g')
plt.stem(ts, xq)
```

```python
error = xs - xq
print(error)

plt.stem(error)

promedio = error.mean()
print(promedio)
```
